#### - 12 panel plot of each month's climatology

#### - Plotted two different ways

```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import pandas as pd
```

```python
file='/homes/swhitak9/clim680data/monthly_precip_India_total.nc'

ds=xr.open_dataset(file, decode_times=False)

ds
```

```python
ds['time']=pd.date_range(start='01-01-1901',end='12-31-2019',freq='M')
ds['time']
```

```python
ds_climo2=ds.groupby('time.month').mean()
```

```python
fig = plt.figure(figsize=(13,10.5))


for i,month in enumerate(ds_climo2['month']):
    plt.subplot(3,4,i+1)
    cs=plt.contourf(ds_climo2['pr'].sel(month=month), cmap='BrBG')
    plt.title(month.values)
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    
plt.suptitle('Monthly Precipitation Climatology over India')
    

fig.subplots_adjust(wspace=0.4,hspace=0.5)


cbar_ax=fig.add_axes([0.2,0.05,0.6,0.02])  


cbar=fig.colorbar(cs,cax=cbar_ax,orientation='horizontal',label='mm/year')
```
